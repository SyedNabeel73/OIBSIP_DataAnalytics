```python
#Handle missing values
init_rows = len(df)
df.dropna(inplace=True)
rows_after_dropna = len(df)
print(f"Initial rows: {init_rows}")
print(f"Rows after dropping NaNs: {rows_after_dropna}")
print(f"Rows dropped: {init_rows - rows_after_dropna}")
print("\n" + "="*50 + "\n")
```

```python
#cleaning and conveting Installs coloumn
df['Installs'] = df['Installs'].str.replace('+', '', regex=False)
df['Installs'] = df['Installs'].str.replace(',', '', regex=False)
```

```python

df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce').astype(pd.Int64Dtype())
print("Data type for 'Installs' is now:", df['Installs'].dtype)
print("\n" + "="*50 + "\n")
```

```python
#cleaning and converting price coloumn
df['Price'] = df['Price'].str.replace('$', '', regex=False)
# Convert to float
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
print("Data type for 'Price' is now:", df['Price'].dtype)
print("\n" + "="*50 + "\n")
```

```python
df.sample(5)
```

```python
#A.Correlation Analysis 
print("Correlation Matrix (Key Metrics):\n")
numeric_cols = ['Rating', 'Reviews', 'Size', 'Installs', 'Price']
print(df[numeric_cols].corr())
#B.Top 10 Categories by Average Installs (Popularity)
top_categories_installs = df.groupby('Category')['Installs'].mean().sort_values(ascending=False).head(10)
print("\nTop 10 Categories by Average Installs (Popularity):\n", top_categories_installs)
#C.Average Rating by Content Rating
avg_rating_by_content = df.groupby('Content Rating')['Rating'].mean().sort_values(ascending=False)
print("\nAverage Rating by Content Rating:\n", avg_rating_by_content)
```

```python
import plotly.express as px
```

```python
# 3.Distribution of App Ratings
fig_rating = px.histogram(
    df, 
    x='Rating', 
    title='Distribution of App Ratings',
    marginal='box', # Show box plot for distribution detail (Skill Enhancement: Adding detail)
    color_discrete_sequence=['blue']
)
fig_rating.show()
```

```python
#Installs (Popularity) by Category
#Group by category and sum installs (using log scale for better visualization)
category_installs = df.groupby('Category')['Installs'].sum().reset_index()

fig_installs = px.bar(
    category_installs.sort_values('Installs', ascending=False), 
    x='Category', 
    y='Installs', 
    title='Total Installs by App Category (Popularity Trend)',
    log_y=True, # Use log scale to show wide install range clearly (Skill Enhancement: Appropriate scaling)
    labels={'Installs': 'Total Installs (Log Scale)'},
    color='Installs',
    color_continuous_scale=px.colors.sequential.Teal
)
fig_installs.update_layout(xaxis={'categoryorder':'total descending'})
fig_installs.show()
```
