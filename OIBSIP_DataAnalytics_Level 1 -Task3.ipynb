```python
#project 3 of level 1:
#idea:Data Cleaning 
import pandas as pd
dataset = r"C:\Users\syedn\Desktop\EDA_resales _data\AB_NYC_2019.csv"
nyc_data = pd.read_csv(dataset)
nyc_data 
```

```python
#Data Integrity checks
print(nyc_data.info())
print(nyc_data.describe(include='all'))

# Normalize column names by removing extra spaces
nyc_data.columns = nyc_data.columns.str.strip()
```

```python
#Dropping duplicates
#Drop fully duplicated rows
nyc_data.drop_duplicates(inplace=True)
#Ensure listing IDs are unique
nyc_data.drop_duplicates(subset='id',keep='first',inplace=True)
```

```python
#Handling missing data
#Fill missing listing names with a default placeholder
nyc_data['name'] = nyc_data['name'].fillna("Unnamed Listing")
nyc_data['host_name'] = nyc_data['host_name'].fillna("Unknown Host")
nyc_data['last_review'] = pd.to_datetime(nyc_data['last_review'],errors='coerce')
#Fill missing dates with the string "0"
nyc_data['last_review'] = nyc_data['last_review'].fillna("0")
```

```python
#Fill 'reviews_per_month'
#If number_of_reviews is 0, reviews_per_month is logically 0
nyc_data.loc[nyc_data['number_of_reviews'] == 0, 'reviews_per_month'] = 0
# Use neighborhood-specific median, or fall back to global median
global_median = nyc_data['reviews_per_month'].median()
neigh_median = nyc_data.groupby('neighbourhood')['reviews_per_month'].transform('median')
nyc_data['reviews_per_month'] = nyc_data['reviews_per_month'].fillna(neigh_median).fillna(global_median)
```

```python
#Standardization
#Strip leading/trailing spaces from all text fields
for col in nyc_data.select_dtypes(include='object').columns:
    nyc_data[col] = nyc_data[col].astype(str).str.strip()
#Normalize bad text values (like "nan", "None")
nyc_data.replace(to_replace=["nan","NaN","None"],value=pd.NA,inplace=True)
#Reset index for cleanliness
nyc_data.reset_index(drop=True,inplace=True)
```

```python
#Outlier detection
#Remove rows with zero or negative price
nyc_data = nyc_data[nyc_data['price']>0].copy()
#Cap high prices at the 99th percentile
price_cap = nyc_data['price'].quantile(0.99)
nyc_data['price'] = nyc_data['price'].clip(upper=price_cap)
#Cap minimum_nights at 99th percentile
min_nights_cap = nyc_data['minimum_nights'].quantile(0.99)
nyc_data['minimum_nights'] = nyc_data['minimum_nights'].clip(lower=1,upper=min_nights_cap).astype(int)
```
